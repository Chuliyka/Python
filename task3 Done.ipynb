## Додаткове завдання
Створіть генератор рандомних книжок. Книжки мають мати рандомну кількість сторінок, кожна з яких наповнюється з рандомного реєстру слів. Слова об'єднюйте у речення за допомогою крапок і абзаців. Додайте можливість передивлятись книжки у консолі.

```python
class ScientificBook:
    def __init__(self,title,author, literature, glossary):
        self.title = title
        self.author = author
        self.literature = literature
        self.glossary = glossary

    def show_info(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")

        print("literature:")
        for item in self.literature:
            print(f" - {item}")

        print("glossary: ")
        for term, definition in self.glossary.items():
            print(f" - {term}:{definition}")


class Roman:
    def __init__(self, list_of_characters, title, author):
        self.title = title
        self.author = author
        self.list_of_characters = list_of_characters

    def show_info(self):
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print("Characters:")
        for character, descr in self.list_of_characters.items():
            print(f" - {character}: {descr}")


class Guide:
    def __init__(self, img, title, author):
        self.title = title
        self.author = author
        self.img = img

    def show_info(self):
        print(f"title - {self.title}")
        print(f"author - {self.author}")
        print(f"image - {self.img}")


class PageRegistry:
    _instance = None

    def __init__(self):
        self.page_id_counter = 0
        self.page_registry = {}

    def next_pg(self):
        self.page_id_counter += 1
        return self.page_id_counter

    def register_pg(self, page_id, page_content):
        self.page_registry[page_id] = page_content

    def page_content(self, page_id):
        return self.page_registry.get(page_id)

    @classmethod
    def get_instance(cls):
        if cls._instance is None:
            cls._instance = cls()
        return cls._instance


class PageBuilder:
    def build_page(self, content):
        page_id = PageRegistry.get_instance().next_pg()
        PageRegistry.get_instance().register_pg(page_id, content)
        return page_id


class BuildBook:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.format = "Default format"
        self.page_builder = PageBuilder
        self.page_register = PageRegistry.get_instance()

    def format(self, book_format):
        self.format = book_format

    def pg_add(self, content):
        page_id = self.page_builder.build_page(content)
        return page_id

    def book_getter(self):
        return Book(self.author, self.title, self.format, self.page_register)


class Book:
    def __init__(self, title, author, book_format, page_register):
        self.title = title
        self.author = author
        self.format = book_format
        self.page_register = page_register

    def content_getter(self, page_id):
        return self.page_register.content_getter(page_id)

    def show_info(self):
        print(f"format - {self.format}")
        print(f"title - {self.title}")
        print(f"author - {self.author}")


user_choise = input("Яку книгу ви хочете переглянути?\nНаукову?\nРоман?\nПосібник?\nВаш вибір: ")

if user_choise.lower() == "наукову":
    literature = ["Book A", "Book B","Book C"]
    glossary = {"Term 1": "Definition 1", "Term 2": "Definition 2"}

    science_book = ScientificBook("Science", "Ivanka Chuliy", literature,glossary)
    science_book.show_info()
elif user_choise.lower() == "роман":
    list_of_characters = {
        "Jordan": "Cool boy",
        "Bob": "Also cool boy",
        "Max": "Random guy"
    }

    roman = Roman(list_of_characters,"Ivanka's story","Ivanka")
    roman.show_info()

elif user_choise.lower() == "посібник":
    image_path = "padlock.ico"

    guide = Guide(image_path,"Images ITSTEP","John street fotographer")
    guide.show_info()
else:
    print("No books =(")
```
